##### Loading JSON file 

```python
with open('newyork_data.json') as json_data:
    newyork_data = json.load(json_data)
```

###### we will define a new variable to store the data of neighborhood ,all data is in features key


```python
boroughs_data = newyork_data['features']
# extracting the values of list on position 1 
boroughs_data[1]
```

##### Tranforming data to Dataframe


```python
# my dataframe columns will be below
column_names = ['Borough', 'Neighborhood', 'Latitude', 'Longitude'] 

#our dataframe
neighborhood_data = pd.DataFrame(columns=column_names)
neighborhood_data
```

```python
# storing the data from json file to data frame using for loop
for data in boroughs_data:
    borough = neighborhood_name = data['properties']['borough'] 
    neighborhood_name = data['properties']['name']
        
    neighborhood_latlon = data['geometry']['coordinates']
    neighborhood_lat = neighborhood_latlon[1]
    neighborhood_lon = neighborhood_latlon[0]
    
    neighborhood_data = neighborhood_data.append({'Borough': borough,
                                          'Neighborhood': neighborhood_name,
                                          'Latitude': neighborhood_lat,
                                          'Longitude': neighborhood_lon}, ignore_index=True)
    
```

```python
# Checking the number of neighborhoods and rows
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(neighborhood_data['Borough'].unique()),
        neighborhood_data.shape[0]
    )
)
```

#### Creating Map of NYC Using Follium 


```python
# create map of NYC using latitude and longitude values
nyc_map = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhood_data['Latitude'], neighborhood_data['Longitude'], 
                                           neighborhood_data['Borough'], neighborhood_data['Neighborhood']):
    folium.features.CircleMarker(
        [lat, lng],
        radius=5,
        color='red',
        fill = True,
        fill_color='red',
        fill_opacity=0.6
    ).add_to(nyc_map)
    # display map
    
nyc_map 
```
